```python
# build an svg file with a 3d graph containing X, Y, and Z axes, and a surface defined by the function z = f(x, y) = sin(sqrt(x^2 + y^2)).
import numpy as np #import library to perform mathematical operations
import plotly.graph_objects as go #import library to create 3D graphs
import pandas as pd  # import library to read csv files
import seaborn as sns  # import library for data visualization
# import the csv module to read the data from a csv file

# Import the csv file
df = pd.read_csv("C:\\Users\\linhn\\Downloads\\Fear\\Test 35.csv")
df = df.sort_values(by="ID")
# Display the first few rows of the dataframe to verify it was loaded correctly
df.head(5)


# Extract the x, y, and z values from the dataframe
x = df['Imminence'].values
y = df['ID'].values
z = df['CDA.AmpSum'].values

# Create a 3D scatter plot with lines connecting the points
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='lines+markers',
    marker=dict(
        size=4,
        color='white',
        opacity=0.8
    ),
    line=dict(
        color='darkgray',
        width=2
    )
)])

# Apply all layout and styling
fig.update_layout(
    title='3D Data Visualization',
    autosize=False,
    width=1000, # Set width to 1920 pixels later
    height=780, # Set height to 1080 pixels later
    # Set black background for the entire plot
    paper_bgcolor='black',
    plot_bgcolor='black',
    #set default home view
    scene_camera=dict(
        eye=dict(x=3, y=1.5, z=0)
    ),
    # set default font color to white
    font=dict(
        color='white',
        family='Helvetica, sans-serif'
    ),
    # Set axis titles
    scene=dict(
        xaxis_title='Imminence',
        yaxis_title='ID',
        zaxis_title='Overall intensity of reaction (CDA.AmpSum)',
        aspectratio=dict(x=.6, y=1.8, z=1),  # Make ID (X) axis 3x wider
        # Black background
        xaxis=dict(
            showbackground=True,
            backgroundcolor='black',
            showgrid=True,
            gridcolor='gray',
            gridwidth=10,
            zerolinecolor='white'
        ),
        yaxis=dict(
            showbackground=True,
            backgroundcolor='black',
            showgrid=True,
            gridcolor='gray',
            gridwidth=3,
            zerolinecolor='white',
            #show all ID values on the Y axis, but just its number, not the ID label
            tickmode='linear',
            tick0=0,
            dtick=1
        ),
        zaxis=dict(
            showbackground=True,
            backgroundcolor='black',
            showgrid=True,
            gridcolor='gray',
            gridwidth=3,
            zerolinecolor='white'
        )
    )
)

# Export and show
fig.write_html("3d_surface_plot.html")
fig.show()
```
